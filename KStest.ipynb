```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt

import pandas as pd
from pandas import DataFrame
```

```python
# read input data file
df = pd.read_excel('KStest.xlsx')
df
```

```python
list(df.columns.values)
```

```python
#select the comparison data
asy = 'A 10^-7'
gen = 'A 10^-8'

title = asy + ' vs ' + gen
asyFF = asy + ' FF'
genFF = gen + ' FF'
print(title)
```

```python
a = df[asy]
b = a.dropna()
c=b * -1
```

```python
d = df[asyFF]
e = d.dropna()
```

```python
o = df[gen]
p = o.dropna()
q = p * -1
```

```python
r = df[genFF]
s = r.dropna()
```

```python
stats.ks_2samp(c, q)
```

```python

plt.figure(figsize=(8.5, 6.5), dpi=100)
plt.suptitle(title,fontsize=12)

plt.subplot(2,2,1)

bins = np.arange(np.floor(c.min()),np.ceil(c.max()))
plt.hist(c, bins=bins, density=True, histtype="step", color='red')
plt.xlabel('Temperature (-Â°C)')
# plt.ylabel('ylabel')

plt.subplot(2,2,2)
plt.step(c,e, color='red')
plt.xlabel('Temperature (-Â°C)')
plt.ylim((0,1.1))
#plt.ylabel('ylabel')

plt.subplot(2,2,3)

bins = np.arange(np.floor(q.min()),np.ceil(q.max()))
plt.hist(q, bins=bins, density=True, histtype="step", color='blue')
plt.xlabel('Temperature (-Â°C)')
#plt.ylabel('ylabel')

plt.subplot(2,2,4)
plt.step(q,s, color='blue')
plt.xlabel('Temperature (-Â°C)')
plt.ylim((0,1.1))
#plt.ylabel('ylabel')

plt.tight_layout()
plt.subplots_adjust(top=0.94)

plt.savefig('result1.eps', format='eps', dpi=1000)
plt.show()
```

```python

fig = plt.figure()
fig = plt.figure(figsize=(8.5, 3.5), dpi=100)

ax = fig.add_subplot(1,2,1)


bins = np.arange(np.floor(c.min()),np.ceil(c.max()))
ax.hist(c, bins=bins, density=True, alpha=0.5, color='red')
ax.set_xlabel('Temperature (-Â°C)')
# plt.ylabel('ylabel')


bins = np.arange(np.floor(q.min()),np.ceil(q.max()))
ax.hist(q, bins=bins, density=True, alpha=0.5, color='blue')
ax.set_xlabel('Temperature (-Â°C)')
#plt.ylabel('ylabel')


ax2 = fig.add_subplot(1,2,2)

ax2.step(c,e, color='red')
ax2.set_xlabel('Temperature (-Â°C)')
ax2.set_ylim(0,1.1)
#plt.ylabel('ylabel')

ax2.step(q,s, color='blue')
ax2.set_xlabel('Temperature (-Â°C)')
ax2.set_ylim(0,1.1)
#plt.ylabel('ylabel')


fig.savefig('result2.jpg', dpi=1000)

fig.show()
```
